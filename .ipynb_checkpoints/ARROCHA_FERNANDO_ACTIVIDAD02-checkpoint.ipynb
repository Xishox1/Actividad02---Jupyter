<h1 align='center'> Ejercicio 1</h1>

```python
#Lee el fichero de datos tarjetas.txt y almacena la información en un dataframe de Pandas (llámalo df) y comprueba los registros que tiene el archivo.
import pandas as pd
df = pd.read_csv('tarjetas.txt', delimiter='\t', decimal=',')
print(df.shape)
df.info()
#¿Cuántos clientes distintos existen?
clientes_distintos = df["idcliente"].nunique()
print("\nClientes distintos:",clientes_distintos)
#¿Qué cliente es el que tiene un mayor número contratos? Indica su idcliente, en caso de existir más de uno indica todos los id. ¿Cuántos contratos tiene/n este/os cliente/s?
contratos_por_cliente = df["idcliente"].value_counts()
maximo_contratos = contratos_por_cliente.max()
clientes_con_maximo_contratos = contratos_por_cliente[contratos_por_cliente == maximo_contratos].index
print("Los clientes con mas contratos son los clientes con id: ",clientes_con_maximo_contratos,"con un total de ", maximo_contratos, "contratos")
#¿Cuántos contratos hay de que cada tipo de producto?
contratos_por_producto = df["producto"].value_counts()
print("\nContratos por cada tipo de producto:")
print(contratos_por_producto)
#Crea una nueva variable llamada contacto_str en que se haya sustituido sus valores por los siguientes textos:
#Valor 0: rechazado
#Valor 1: ausente
#Valor 2: contactado
# Mapeo de valores originales a nuevos valores
valores_nuevos = {0:"rechazado",1:"ausente",2:"contactado"}

# Crear una nueva columna en el dataframe con los valores mapeados
df["contacto_str"] = df["contacto_cliente"].replace(valores_nuevos)

#Crea una nueva variable llamada pct_uso que indique el porcentaje de utilización calculado como el cociente de dispuesto y limite en tantos por uno.
df["pct_uso"] = (df["dispuesto"] / df["limite"]) * 100

#¿Cúantas tarjetas hay que se encuentren en situciacón de sobredisposición, es decir, que tengan un porcentaje de uso superior a 1?
sobredisposicion = df["pct_uso"].where(df["pct_uso"]>100)
print("\nTarjetas en sobredisposición: ",sobredisposicion.count())

#¿Cuál es el límite mínimo, máximo y medio por producto? Muéstralo en un único dataframe.
# Agrupando los datos por producto y aplicando los métodos min, max y mean
#df_limites = df.groupby("producto")["limite"].agg(["min", "max","mean"]).reset_index()
df_limites = df.groupby("producto")["limite"].agg(['min','max','mean'])
df_limites = df_limites.rename(columns={'min':'minimo', 'max':'maximo', 'mean':'medio'})
df_limites = df_limites.sort_values(by='producto')
print("\nLímite mínimo, máximo y medio por producto:\n\n",df_limites)
```

```python
#Ya que existe un número determinado de clientes que han rechazado el contacto, genera un nuevo dataframe sin 
#estos clientes llamado df_fil que sea una copia de df.
df_fil = df.loc[df["contacto_cliente"]!=0].copy()
#Por cuestiones de calidad se ha detectado que existen contratos con un tipo de interés que no es uno de los estipulados para cada producto. Sobre df_fil genera una nueva variable llamada tipo_corr que tenga el tipo de interés correcto para cada producto ¿cuántos casos con tipo incorrecto hay?
#diccionario con los tipos de interés correctos para cada producto
tipos = {'A': 18.75, 'B': 10.25, 'C': 6.33}

df_fil["tipo_corr"] = df_fil["producto"].apply(lambda x: tipos[x])

casos_tipo_incorrecto = df_fil[df_fil["tipo"] != df_fil["tipo_corr"]].shape[0]
print("Existen ", casos_tipo_incorrecto, " casos con tipo de interés incorrecto.")

#Sobre df_fil realiza las siguientes comprobaciones:
#•	¿Sobre los clientes contactados cual es el producto mayoritario?
df_fil_contactados = df_fil.loc[df_fil["contacto_cliente"]==2]
producto_mayoritario = df_fil_contactados["producto"].value_counts().idxmax()
print("El producto mayoritario entre los clientes contactados es: ", producto_mayoritario)
#•	¿que porcentaje de tarjetas se encuentran en situación de Impago?
total_tarjetas = df_fil.shape[0]
tarjetas_impago = df_fil.loc[df_fil["impago"]==1].shape[0]
porcentaje_impago = (tarjetas_impago/total_tarjetas)*100
print("El porcentaje de tarjetas en situación de impago es: {:.2f}%".format(porcentaje_impago))
#•	Se considera que la cartera se encuentra en situación sostenible si el porcentaje de tarjetas en situación contable dudoso y normal supera el 80% del total de tarjetas ¿se verifica esta condición?
total_tarjetas = df_fil.shape[0]
tarjetas_dudoso_normal = df_fil.loc[df_fil["impago"].isin([2,3])].shape[0]
porcentaje_dudoso_normal = (tarjetas_dudoso_normal/total_tarjetas)*100

if porcentaje_dudoso_normal > 80:
    print("La cartera se encuentra en situación sostenible. El porcentaje de tarjetas en situación dudoso y normal es del {:.2f}%".format(porcentaje_dudoso_normal))
else:
    print("La cartera no se encuentra en situación sostenible. El porcentaje de tarjetas en situación dudoso y normal es del {:.2f}%".format(porcentaje_dudoso_normal))
#En este caso se utiliza el método .isin() para seleccionar solo aquellas filas donde el valor de impago es 2 o 3.
#El valor impreso sería el porcentaje de tarjetas en situación de dudoso y normal en el dataframe df_fil y se verificaria si cumplen con la condición

#•	Para la empresa el importe dispuesto de cada tarjeta genera un coste que depende de la situación contable:
#	Fallido: 120%.
#	Impago: 100%.
#	Dudoso: 10%
#	Normal: 0%
#¿Cual es el coste total que supone a la empresa esta cartera?
costes = {0:1.2,1:1,2:0.1,3:0}
# Me marcó un error de MemoryError y con la línea 43 se reduce el tamaño del dataframe y se evita el error MemoryError
df_fil["coste"] = df_fil.apply(lambda x: x["dispuesto"]*costes[x["impago"]], axis=1)
coste_total = df_fil["coste"].sum()
print("El coste total que supone a la empresa esta cartera es: {:.2f}".format(coste_total))

#Cada tarjeta trae asociado un coste para el titular de la tarjeta, existen varios tipos de costes que se detallan a continuación:
#•	Coste por disposición: Es el coste que se calcula sobre el importe dispuesto, solo se aplica si se tiene un porcentaje dispuesto superior al 50% e inferior o igual al 100% y depende del producto, los porcentajes a aplicar son:
#	1% para el producto A.
#	3% para el producto B.
#	0% para el producto C.
#•	Coste por sobre disposición: Es el coste que se calcula sobre el importe dispuesto por encima del límite:
#	15% para el producto: A.
#	10% para el producto: B.
#	5% para el producto: C.
#•	Coste de no disposición: El coste de no disposición es el aplicado sobre el importe disponible siempre que el porcentaje de uso sea inferior al 20%. Y es:
#	2% para las tarjetas sin importes dispuestos.
#	Para el resto de tarjetas un 0,15% por cada 5% dispuesto, es decir si es el porcentaje dispuesto está entre (0-5%] se asigna un 0,15% para ese tramo, si (5%-10%] se asigna un 0,30% un para ese tramo, etc.
#•	Coste global de mantenimiento: Este coste se asigna proporcionalmente según el porcentaje de uso de la tarjeta. Aplicando una comisión mínima del 1,23% sobre el límite de la tarjeta y un máximo del 3%.
#Teniendo en cuenta lo anterior, crea una columna en df_fil que contenga el coste total por tarjeta, llámala imp_comisiones.

def coste_tarjeta(row):
    coste_disposicion = 0
    coste_sobre_disposicion = 0
    coste_no_disposicion = 0
    coste_mantenimiento = 0
    pct_uso = row["dispuesto"]/row["limite"]*100
    producto = row["producto"]
    if producto == "A":
        if pct_uso > 50 and pct_uso <=100:
            coste_disposicion = row["dispuesto"]*0.01
        if pct_uso > 100:
            coste_sobre_disposicion = (pct_uso - 100)*0.15
    elif producto == "B":
        if pct_uso > 50 and pct_uso <=100:
            coste_disposicion = row["dispuesto"]*0.03
        if pct_uso > 100:
            coste_sobre_disposicion = (pct_uso - 100)*0.10
    else:
        if pct_uso > 50 and pct_uso <=100:
            coste_disposicion = 0
        if pct_uso > 100:
            coste_sobre_disposicion = (pct_uso - 100)*0.05
    if pct_uso < 20:
        coste_no_disposicion = row["disponible"]*0.02
    coste_mantenimiento = (1.23 + (pct_uso/100)*(3-1.23))*row["limite"]
    return coste_disposicion + coste_sobre_disposicion + coste_no_disposicion + coste_mantenimiento

   
df_fil["imp_comisiones"] = df_fil.apply(coste_tarjeta, axis=1)

#Crea un dataframe nuevo llamado df_coste_cliente que contenga para cada cliente el coste total de sus tarjetas 
#(imp_comisiones) ¿Alguno de los clientes con mayor número de tarjetas (el obtenido en el ejercicio 2.1) 
#es el que paga mayor importe en comisiones?


df_coste_cliente = df_fil.groupby(["idcliente"])["imp_comisiones"].sum().reset_index()
#ordeno el dataframe df_coste_cliente para conocer cual es el cliente que paga mayor importe en comisiones.
df_coste_cliente = df_coste_cliente.sort_values("imp_comisiones", ascending = False)
#variable llamada df_contratos que contenga la información de los contratos agrupados por cliente con el número de contratos y el id del cliente
df_contratos = df_fil.groupby(["idcliente"]).size().reset_index(name="n_contratos")
#Se puede comparar los idcliente del dataframe df_coste_cliente con los obtenidos en el ejercicio 1 (en el que se identificó al cliente con mayor número de contratos) para verificar si es el mismo o no el que paga mayor importe en comisiones.
idcliente_mas_contratos = df_contratos.loc[df_contratos["n_contratos"] == df_contratos["n_contratos"].max(), "idcliente"]
coste_cliente_mas_contratos = df_coste_cliente.loc[df_coste_cliente["idcliente"].isin(idcliente_mas_contratos), "imp_comisiones"]

#imprimir el coste total de comisiones de ese cliente y compararlo con el total de los demás clientes para verificar si es el que más paga en comisiones.
print("El coste total de comisiones del cliente con mas contratos es: {:.2f}".format(coste_cliente_mas_contratos.sum()))
print("El promedio del coste total de comisiones de los demás clientes es: {:.2f}".format(df_coste_cliente.loc[~df_coste_cliente["idcliente"].isin(idcliente_mas_contratos),"imp_comisiones"].mean()))


#                                EJERCICIO 3
#Por cuestiones comerciales se han seleccionado cinco clientes: 12874803, 13067217, 13163055, 13335511, 12900036. Genera un dataframe llamado clientes_vip que contenta los siguientes campos e información para cada cliente:
#•	Edad: [41, 18, 33, 56, 42]
#•	Nombre: [Pepa A., Antonio G., Fernando T., Rebeca S., Laura F.]
#•	Ingresos anuales: [43000, 25000, 55000, 32000, 40000]
import pandas as pd

clientes_vip = pd.DataFrame({"idcliente": [12874803, 13067217, 13163055, 13335511, 12900036], 
                            "Edad": [41, 18, 33, 56, 42], 
                            "Nombre": ["Pepa A.", "Antonio G.", "Fernando T.", "Rebeca S.", "Laura F."],
                            "Ingresos anuales": [43000, 25000, 55000, 32000, 40000]})
#Genera un nuevo dataframe llamado cli_vip_fin que sea el resultado de añadir la información de clientes_vip a df_fil y que contenga únicamente los cinco clientes seleccionados.
cli_vip_fin = pd.merge(df_fil, clientes_vip, on='idcliente')
cli_vip_fin["bonificacion"] = cli_vip_fin["Edad"].apply(lambda x: x//10)
print("\n\t******** Mostrando Dataframe del ejercicio 3 **********\n")
print(cli_vip_fin)

```

```python
#Realiza una copia del objeto df y llámalo df2_4, sobre este conjunto de datos realiza las siguientes operaciones:
#Crea una variable llamada limite_gr de forma que contenga una versión categorizada de la variable limite conforme a los siguientes tramos:

#[0, 700]
#(700, 800]
#(800, 1200]
#(1200, 1700]
#(1700, 2500]
#(2500, 5000]
#(5000, 8000]
#(8000, 10000]
#(10000, 20000]
import pandas as pd
import numpy as np
df = pd.read_csv('tarjetas.txt', delimiter='\t', decimal=',')
df2_4 = df.copy()
bins = [0, 700, 800, 1200, 1700, 2500, 5000, 8000, 10000, 20000]
labels = ["[0, 700]", "(700, 800]", "(800, 1200]", "(1200, 1700]", "(1700, 2500]", "(2500, 5000]", "(5000, 8000]", "(8000, 10000]", "(10000, 20000]"]
df2_4["limite_gr"] = pd.cut(df2_4["limite"], bins, labels=labels)

#¿Cuántos registros hay en cada categoría de la variable limite_gr? ¿Cuál es el que acumula un mayor porcentaje de registros y qué porcentaje es?
value_counts = df2_4['limite_gr'].value_counts()
value_counts_pct = df2_4['limite_gr'].value_counts(normalize=True) * 100

print("Registros en cada categoría\n")
print(value_counts)
print("\nPorcentaje en cada categoría\n")
print(value_counts_pct)

percentages = value_counts / len(df2_4) * 100
max_category = percentages.idxmax()
max_percentage = percentages.max()

print("\nLa categoría con el porcentaje más alto es " + max_category + " con " + str(max_percentage) + "%")

#Selecciona una muestra aleatoria del 10% de los registros de df2_4 y almacenála en df2_4_smp. 
#¿Qué porcentaje de registros hay en cada categoría de la variable limite_gr en la muestra?
df2_4_smp = df2_4.sample(frac=0.1, random_state=1)
value_counts_smp = df2_4_smp['limite_gr'].value_counts(normalize=True) * 100
print("---------------------------------------------------------------------------------")
print("Muestra aleatoria del 10% con sus porcentajes de registro en cada categoria de la variable limite_gr\n",value_counts_smp)

# Compara ambos porcentajes en el mismo dataframe.
#Calculo los porcentajes en la muestra
df2_4_smp_pct = df2_4_smp['limite_gr'].value_counts(normalize=True) * 100

#Combino ambas tablas de porcentajes
df_pct = pd.concat([value_counts_pct, df2_4_smp_pct], axis=1)
print("\n\t\t**COMBINANDO AMBAS TABLAS DE PORCENTAJES**")
df_pct.columns = ['Porcentaje en todos los registros', 'Porcentaje en la muestra']

#Imprimo el dataframe resultante
print(df_pct)

#Dentro de la empresa se utiliza la cifra
#𝑑𝑖𝑠𝑝𝑜𝑛𝑖𝑏𝑙𝑒2−𝑘⋅𝑑𝑖𝑠𝑝𝑢𝑒𝑠𝑡𝑜𝑙𝑖𝑚𝑖𝑡𝑒2

#como un factor de propensión a la sobreutilización de la tarjeta. El componente  𝑘  toma un valor que depende del límite de la operación:
#k = 0.5 si limite < 1500
#k = 1 si limite < 5000
#k = 1.2 si limite < 10.000
#k = 1.5 si limite >= 10.000 Crea una columna llamada propension de forma que contenga el cálculo anterior realizado a través de una función.

import matplotlib.pyplot as plt
def calcula_propension(row):
    if row['limite'] < 1500:
        k = 0.5
    elif row['limite'] < 5000:
        k = 1
    elif row['limite'] < 10000:
        k = 1.2
    else:
        k = 1.5
    propension = (((row['disponible']**2 - k)*(row['dispuesto']))/(row['limite']**2))
    return propension

df2_4['propension'] = df2_4.apply(calcula_propension, axis=1)

#Dibuja el histograma de la variable propension.
plt.hist(df2_4["propension"])
plt.xlabel("Propension")
plt.ylabel("Frecuencia")
plt.title("Histograma de Propension")
plt.show()


#¿Cuántos registros tienen una propensión entre 0.95 y 1.05?
registros_entre_0_95_y_1_05 = df2_4.loc[(df2_4["propension"] > 0.95) & (df2_4["propension"] < 1.05)].shape[0]
print("Registros que tienen una propension entre 0.95 y 1.05: ",registros_entre_0_95_y_1_05)

# ¿Cada factor  𝑘  a cuántos registros ha sido aplicado?
def asigna_k(limite):
    if limite < 1500:
        return 0.5
    elif limite < 5000:
        return 1
    elif limite < 10000:
        return 1.2
    else:
        return 1.5

df2_4["k"] = df2_4["limite"].apply(asigna_k)

counts_k = df2_4["k"].value_counts()
print("\n***cantidad de registros tienen cada valor de k***")
print(counts_k)

# La propension permite conocer si hay un alto o bajo riesgo de que el limite de la tarjeta pueda llegar a ser superado.
#Para ello es necesario identificar que registros son de alto riesgo y cuáles no. 
#Crea una columna llamada riesgo que contenga los valores alto riesgo y bajo riesgo, 
#se considera que un cliente es de alto riesgo si la propensión supera al porcentaje de uso.

def evaluar_riesgo(propension, porcentaje_uso):
    if propension > porcentaje_uso:
        return "alto riesgo"
    else:
        return "bajo riesgo"

porcentaje_uso = 0.8
df2_4["riesgo"] = df2_4["propension"].apply(lambda x: evaluar_riesgo(x, porcentaje_uso))

#Para priorizar acciones comerciales se desea marcar al 5% de los clientes con menor porcentaje de uso con la etiqueta 
#prioridad 1, al 1% de los clientes con mayor porcentaje de uso como prioridad 4, al 21% de los clientes con
#mayor porcentaje de uso con prioridad 3 y el resto con prioridad 2.
#El porcentaje de uso a utilizar debe ser agregado para cada cliente ¿cuántos clientes hay en cada grupo de prioridad?
df2_4['porcentaje_uso'] = (df2_4['disponible'] / df2_4['limite']) * 100

bins = [0, df2_4['porcentaje_uso'].quantile(.05),
        df2_4['porcentaje_uso'].quantile(.96),
        df2_4['porcentaje_uso'].quantile(.99),
        100]
labels = ['prioridad 1', 'prioridad 2', 'prioridad 3', 'prioridad 4']

df2_4['prioridad'] = pd.cut(df2_4['porcentaje_uso'], bins=bins, labels=labels)
print("\n***Cantidad de clientes en cada grupo de prioridad***")
print(df2_4['prioridad'].value_counts())


#La empresa ha decidido aplicar una comisión de un 1,5% a las tarjetas de alto riesgo sobre el importe disponible y una bonificación del 0,43% a las tarjetas de bajo riesgo sobre el importe dispuesto.
#•	¿Cuál es el importe total recaudado por la empresa con este nuevo importe?

importe_alto_riesgo = df2_4.loc[df2_4['riesgo'] == 'alto riesgo', 'disponible'].sum() * 0.015
importe_bajo_riesgo = df2_4.loc[df2_4['riesgo'] == 'bajo riesgo', 'dispuesto'].sum() * 0.0043
importe_total = importe_alto_riesgo + importe_bajo_riesgo
print("\nEl importe total recaudado por la empresa es: ", importe_total)

# ¿Qué coste tiene para la empresa esta nueva bonificación?
importe_bajo_riesgo = df2_4[df2_4['riesgo'] == 'bajo riesgo']['disponible'].sum()
coste_bonificacion = importe_bajo_riesgo * 0.0043
print("El coste de la bonificación para la empresa es de: ", coste_bonificacion)

# ¿Ha resultado beneficioso para la empresa aplicar esta nueva medida?
importe_comisiones = (df2_4[df2_4['riesgo'] == 'alto riesgo']['disponible'] * 0.015).sum()
importe_bonificaciones = (df2_4[df2_4['riesgo'] == 'bajo riesgo']['dispuesto'] * 0.0043).sum()
beneficio_neto = importe_bonificaciones - importe_comisiones

if beneficio_neto > 0:
    print("La nueva medida ha resultado beneficiosa para la empresa, ya que ha obtenido un beneficio neto de: ", beneficio_neto)
else:
    print("La nueva medida no ha resultado beneficiosa para la empresa, ya que ha obtenido un beneficio neto de: ", beneficio_neto)
```

```python
#La empresa se está planteando transformar ciertas tarjetas en préstamos. Para cuando los clientes lo soliciten será necesario disponer del cuadro de amortización resultante de realizar esta transformación. Para ello es necesario lo siguiente:
#•	El importe del préstamo será el saldo dispuesto de la tarjeta +10% de comisión +5% del límite si la tarjeta no está sobredispuesta en este porcentaje.
#•	Se considerará un tipo de interés fijo del 2,75%, aunque debe poder elegirse otro.
#•	El plazo amortización será preestablecido en 5 años, pero podría modificarse.
#•	Se considera sistema de amortización francés (cuota fija) para amortización mensual.

import pandas as pd
import numpy as np
# Datos del préstamo
saldo_dispuesto = 1000  # importe del saldo dispuesto
comision = 0.1   # comisión del 10%
limite = 5000
porcentaje_sobre = 0.05
tasa_interes = 0.0275  # tasa de interés anual del 2.75%
plazo = 60  # plazo en meses

# Cálculo del importe a prestar
importe_prestamo = saldo_dispuesto + saldo_dispuesto * comision
if limite * porcentaje_sobre > importe_prestamo:
    importe_prestamo += limite * porcentaje_sobre

#cálculo del factor de amortización
factor_amortizacion = (tasa_interes * (1 + tasa_interes) ** plazo) / ((1 + tasa_interes) ** plazo - 1)

#Inicialización de listas para almacenar los resultados
cuotas = []
intereses = []
amortizaciones = []
saldos = []

#cálculo de las cuotas mensuales
saldo_restante = importe_prestamo
for i in range(plazo):
    interes = saldo_restante * tasa_interes
    amortizacion = factor_amortizacion - interes
    cuota = interes + amortizacion
    saldo_restante -= amortizacion
    cuotas.append(cuota)
    intereses.append(interes)
    amortizaciones.append(amortizacion)
    saldos.append(saldo_restante)
    
#No hice la función del cuadro de amortizacion porque no me salió, asi que resolví el problema de la siguiente manera:
#Creación del cuadro de amortización
data = {
'cuota': cuotas,
'interes': intereses,
'amortizacion': amortizaciones,
'saldo': saldos
}

df_amortizacion = pd.DataFrame(data)
 

#Imprimir cuadro de amortización
print("***Cuadro de amortización***\n")
print(df_amortizacion)


#Con las condiciones prefijadas de antemano ¿cuál sería la cuota inicial que se pagaría por cada tarjeta?
cuota_inicial = factor_amortizacion * importe_prestamo
print("\nLa cuota inicial a pagar es: ",cuota_inicial)

#Esta operación solo sería aprobada si la cuota resultante supone un importe inferior al límite de la tarjeta dividido entre 55 ¿qué porcentaje de tarjetas es susceptible de ser aprobado?
import random

# límite de la tarjeta
limite = 5000

# número de operaciones a simular
num_operaciones = 1000

# Contador de operaciones aprobadas
operaciones_aprobadas = 0

for i in range(num_operaciones):
    # Generar datos aleatorios para el saldo dispuesto y la comisión
    saldo_dispuesto = random.randint(500,3000)
    comision = random.uniform(0.05,0.15)
    # Cálculo del importe a prestar
    importe_prestamo = saldo_dispuesto + saldo_dispuesto * comision
    #cálculo del factor de amortización
    factor_amortizacion = (tasa_interes * (1 + tasa_interes) ** plazo) / ((1 + tasa_interes) ** plazo - 1)
    #cálculo de la cuota inicial
    interes = importe_prestamo * tasa_interes
    amortizacion = factor_amortizacion - interes
    cuota_inicial = interes + amortizacion
    # Comparar la cuota inicial con el límite de la tarjeta dividido entre 55
    if cuota_inicial <= limite / 55:
        operaciones_aprobadas += 1

# Calculo el porcentaje de operaciones aprobadas
porcentaje_aprobadas = (operaciones_aprobadas / num_operaciones) * 100
print("El porcentaje de operaciones aprobadas es: " + str(round(porcentaje_aprobadas, 2)) + "%")

#El porcentaje de operaciones aprobadas dependerá del rango de los valores aleatorios generados para el saldo dispuesto 
#y la comisión. Empero,se puede inferir que el porcentaje de operaciones aprobadas será menor al 55% ya que 
#la cuota inicial siempre será mayor

```

```python
#Programa una función que permita aplicar la transformación de Box - Cox sobre una columna de un dataframe de Pandas. 
#Selecciona un valor para  𝜆  y utilízalo para aplicar la transformación sobre la variable dispuesto. 
#Aplícale también el valor 0.982.
from scipy.stats import boxcox,jarque_bera
import pandas as pd
import numpy as np
from scipy.optimize import minimize_scalar

def boxcox_transform(dataframe, column, lambda_value):
    dataframe[column] = boxcox(dataframe[column], lambda_value)
    return dataframe

# Uso
df = pd.DataFrame({'x': [1, 2, 3, 4, 5]})
df_transformed = boxcox_transform(df, 'x', 0.982)
print(df_transformed)

#Programa una función que permita obtener el valor  𝜆  óptimo, con una precisiónd de dos decimales, 
#al aplicar la familia de transformaciones Box - Cox sobre la variable dispuesto de forma que la variable 
#se asemeje lo máximo posible a una distribución Normal.

def jb_box_cox(x, lmbda):
    y = (x**lmbda - 1)/lmbda
    return jarque_bera(y)[0]

def encontrar_lambda_optim(x):
    res = minimize_scalar(jb_box_cox, args=(x,), bounds=(-1, 1), method='bounded')
    return round(res.x, 2)


x = np.array((1, 23, 4))
lambda_optim = encontrar_lambda_optim(x)
print("\nValor óptimo de 𝜆: ",lambda_optim)

#Dibuja conjuntamente el histograma de la varible transformada y la variable sin transformar.
import matplotlib.pyplot as plt
y = (x**lambda_optim - 1)/lambda_optim
plt.hist(x, bins=20, color='blue', alpha=0.5, label='x sin transformar')
plt.hist(y, bins=20, color='red', alpha=0.5, label='x transformada') #label es el título que llevará en el gráfico.
plt.title("Histograma de la variable x")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.legend() #añadir leyenda al gráfico 
plt.show() #mostrar el gráfico




```
