<h1 align='center'> Ejercicio 1</h1>

```python
#Lee el fichero de datos tarjetas.txt y almacena la informaci√≥n en un dataframe de Pandas (ll√°malo df) y comprueba los registros que tiene el archivo.
import pandas as pd
df = pd.read_csv('tarjetas.txt', delimiter='\t', decimal=',')
print(df.shape)
df.info()
#¬øCu√°ntos clientes distintos existen?
clientes_distintos = df["idcliente"].nunique()
print("\nClientes distintos:",clientes_distintos)
#¬øQu√© cliente es el que tiene un mayor n√∫mero contratos? Indica su idcliente, en caso de existir m√°s de uno indica todos los id. ¬øCu√°ntos contratos tiene/n este/os cliente/s?
contratos_por_cliente = df["idcliente"].value_counts()
maximo_contratos = contratos_por_cliente.max()
clientes_con_maximo_contratos = contratos_por_cliente[contratos_por_cliente == maximo_contratos].index
print("Los clientes con mas contratos son los clientes con id: ",clientes_con_maximo_contratos,"con un total de ", maximo_contratos, "contratos")
#¬øCu√°ntos contratos hay de que cada tipo de producto?
contratos_por_producto = df["producto"].value_counts()
print("\nContratos por cada tipo de producto:")
print(contratos_por_producto)
#Crea una nueva variable llamada contacto_str en que se haya sustituido sus valores por los siguientes textos:
#Valor 0: rechazado
#Valor 1: ausente
#Valor 2: contactado
# Mapeo de valores originales a nuevos valores
valores_nuevos = {0:"rechazado",1:"ausente",2:"contactado"}

# Crear una nueva columna en el dataframe con los valores mapeados
df["contacto_str"] = df["contacto_cliente"].replace(valores_nuevos)

#Crea una nueva variable llamada pct_uso que indique el porcentaje de utilizaci√≥n calculado como el cociente de dispuesto y limite en tantos por uno.
df["pct_uso"] = (df["dispuesto"] / df["limite"]) * 100

#¬øC√∫antas tarjetas hay que se encuentren en situciac√≥n de sobredisposici√≥n, es decir, que tengan un porcentaje de uso superior a 1?
sobredisposicion = df["pct_uso"].where(df["pct_uso"]>100)
print("\nTarjetas en sobredisposici√≥n: ",sobredisposicion.count())

#¬øCu√°l es el l√≠mite m√≠nimo, m√°ximo y medio por producto? Mu√©stralo en un √∫nico dataframe.
# Agrupando los datos por producto y aplicando los m√©todos min, max y mean
#df_limites = df.groupby("producto")["limite"].agg(["min", "max","mean"]).reset_index()
df_limites = df.groupby("producto")["limite"].agg(['min','max','mean'])
df_limites = df_limites.rename(columns={'min':'minimo', 'max':'maximo', 'mean':'medio'})
df_limites = df_limites.sort_values(by='producto')
print("\nL√≠mite m√≠nimo, m√°ximo y medio por producto:\n\n",df_limites)
```

```python
#Ya que existe un n√∫mero determinado de clientes que han rechazado el contacto, genera un nuevo dataframe sin 
#estos clientes llamado df_fil que sea una copia de df.
df_fil = df.loc[df["contacto_cliente"]!=0].copy()
#Por cuestiones de calidad se ha detectado que existen contratos con un tipo de inter√©s que no es uno de los estipulados para cada producto. Sobre df_fil genera una nueva variable llamada tipo_corr que tenga el tipo de inter√©s correcto para cada producto ¬øcu√°ntos casos con tipo incorrecto hay?
#diccionario con los tipos de inter√©s correctos para cada producto
tipos = {'A': 18.75, 'B': 10.25, 'C': 6.33}

df_fil["tipo_corr"] = df_fil["producto"].apply(lambda x: tipos[x])

casos_tipo_incorrecto = df_fil[df_fil["tipo"] != df_fil["tipo_corr"]].shape[0]
print("Existen ", casos_tipo_incorrecto, " casos con tipo de inter√©s incorrecto.")

#Sobre df_fil realiza las siguientes comprobaciones:
#‚Ä¢	¬øSobre los clientes contactados cual es el producto mayoritario?
df_fil_contactados = df_fil.loc[df_fil["contacto_cliente"]==2]
producto_mayoritario = df_fil_contactados["producto"].value_counts().idxmax()
print("El producto mayoritario entre los clientes contactados es: ", producto_mayoritario)
#‚Ä¢	¬øque porcentaje de tarjetas se encuentran en situaci√≥n de Impago?
total_tarjetas = df_fil.shape[0]
tarjetas_impago = df_fil.loc[df_fil["impago"]==1].shape[0]
porcentaje_impago = (tarjetas_impago/total_tarjetas)*100
print("El porcentaje de tarjetas en situaci√≥n de impago es: {:.2f}%".format(porcentaje_impago))
#‚Ä¢	Se considera que la cartera se encuentra en situaci√≥n sostenible si el porcentaje de tarjetas en situaci√≥n contable dudoso y normal supera el 80% del total de tarjetas ¬øse verifica esta condici√≥n?
total_tarjetas = df_fil.shape[0]
tarjetas_dudoso_normal = df_fil.loc[df_fil["impago"].isin([2,3])].shape[0]
porcentaje_dudoso_normal = (tarjetas_dudoso_normal/total_tarjetas)*100

if porcentaje_dudoso_normal > 80:
    print("La cartera se encuentra en situaci√≥n sostenible. El porcentaje de tarjetas en situaci√≥n dudoso y normal es del {:.2f}%".format(porcentaje_dudoso_normal))
else:
    print("La cartera no se encuentra en situaci√≥n sostenible. El porcentaje de tarjetas en situaci√≥n dudoso y normal es del {:.2f}%".format(porcentaje_dudoso_normal))
#En este caso se utiliza el m√©todo .isin() para seleccionar solo aquellas filas donde el valor de impago es 2 o 3.
#El valor impreso ser√≠a el porcentaje de tarjetas en situaci√≥n de dudoso y normal en el dataframe df_fil y se verificaria si cumplen con la condici√≥n

#‚Ä¢	Para la empresa el importe dispuesto de cada tarjeta genera un coste que depende de la situaci√≥n contable:
#ÔÇß	Fallido: 120%.
#ÔÇß	Impago: 100%.
#ÔÇß	Dudoso: 10%
#ÔÇß	Normal: 0%
#¬øCual es el coste total que supone a la empresa esta cartera?
costes = {0:1.2,1:1,2:0.1,3:0}
# Me marc√≥ un error de MemoryError y con la l√≠nea 43 se reduce el tama√±o del dataframe y se evita el error MemoryError
df_fil["coste"] = df_fil.apply(lambda x: x["dispuesto"]*costes[x["impago"]], axis=1)
coste_total = df_fil["coste"].sum()
print("El coste total que supone a la empresa esta cartera es: {:.2f}".format(coste_total))

#Cada tarjeta trae asociado un coste para el titular de la tarjeta, existen varios tipos de costes que se detallan a continuaci√≥n:
#‚Ä¢	Coste por disposici√≥n: Es el coste que se calcula sobre el importe dispuesto, solo se aplica si se tiene un porcentaje dispuesto superior al 50% e inferior o igual al 100% y depende del producto, los porcentajes a aplicar son:
#ÔÇß	1% para el producto A.
#ÔÇß	3% para el producto B.
#ÔÇß	0% para el producto C.
#‚Ä¢	Coste por sobre disposici√≥n: Es el coste que se calcula sobre el importe dispuesto por encima del l√≠mite:
#ÔÇß	15% para el producto: A.
#ÔÇß	10% para el producto: B.
#ÔÇß	5% para el producto: C.
#‚Ä¢	Coste de no disposici√≥n: El coste de no disposici√≥n es el aplicado sobre el importe disponible siempre que el porcentaje de uso sea inferior al 20%. Y es:
#ÔÇß	2% para las tarjetas sin importes dispuestos.
#ÔÇß	Para el resto de tarjetas un 0,15% por cada 5% dispuesto, es decir si es el porcentaje dispuesto est√° entre (0-5%] se asigna un 0,15% para ese tramo, si (5%-10%] se asigna un 0,30% un para ese tramo, etc.
#‚Ä¢	Coste global de mantenimiento: Este coste se asigna proporcionalmente seg√∫n el porcentaje de uso de la tarjeta. Aplicando una comisi√≥n m√≠nima del 1,23% sobre el l√≠mite de la tarjeta y un m√°ximo del 3%.
#Teniendo en cuenta lo anterior, crea una columna en df_fil que contenga el coste total por tarjeta, ll√°mala imp_comisiones.

def coste_tarjeta(row):
    coste_disposicion = 0
    coste_sobre_disposicion = 0
    coste_no_disposicion = 0
    coste_mantenimiento = 0
    pct_uso = row["dispuesto"]/row["limite"]*100
    producto = row["producto"]
    if producto == "A":
        if pct_uso > 50 and pct_uso <=100:
            coste_disposicion = row["dispuesto"]*0.01
        if pct_uso > 100:
            coste_sobre_disposicion = (pct_uso - 100)*0.15
    elif producto == "B":
        if pct_uso > 50 and pct_uso <=100:
            coste_disposicion = row["dispuesto"]*0.03
        if pct_uso > 100:
            coste_sobre_disposicion = (pct_uso - 100)*0.10
    else:
        if pct_uso > 50 and pct_uso <=100:
            coste_disposicion = 0
        if pct_uso > 100:
            coste_sobre_disposicion = (pct_uso - 100)*0.05
    if pct_uso < 20:
        coste_no_disposicion = row["disponible"]*0.02
    coste_mantenimiento = (1.23 + (pct_uso/100)*(3-1.23))*row["limite"]
    return coste_disposicion + coste_sobre_disposicion + coste_no_disposicion + coste_mantenimiento

   
df_fil["imp_comisiones"] = df_fil.apply(coste_tarjeta, axis=1)

#Crea un dataframe nuevo llamado df_coste_cliente que contenga para cada cliente el coste total de sus tarjetas 
#(imp_comisiones) ¬øAlguno de los clientes con mayor n√∫mero de tarjetas (el obtenido en el ejercicio 2.1) 
#es el que paga mayor importe en comisiones?


df_coste_cliente = df_fil.groupby(["idcliente"])["imp_comisiones"].sum().reset_index()
#ordeno el dataframe df_coste_cliente para conocer cual es el cliente que paga mayor importe en comisiones.
df_coste_cliente = df_coste_cliente.sort_values("imp_comisiones", ascending = False)
#variable llamada df_contratos que contenga la informaci√≥n de los contratos agrupados por cliente con el n√∫mero de contratos y el id del cliente
df_contratos = df_fil.groupby(["idcliente"]).size().reset_index(name="n_contratos")
#Se puede comparar los idcliente del dataframe df_coste_cliente con los obtenidos en el ejercicio 1 (en el que se identific√≥ al cliente con mayor n√∫mero de contratos) para verificar si es el mismo o no el que paga mayor importe en comisiones.
idcliente_mas_contratos = df_contratos.loc[df_contratos["n_contratos"] == df_contratos["n_contratos"].max(), "idcliente"]
coste_cliente_mas_contratos = df_coste_cliente.loc[df_coste_cliente["idcliente"].isin(idcliente_mas_contratos), "imp_comisiones"]

#imprimir el coste total de comisiones de ese cliente y compararlo con el total de los dem√°s clientes para verificar si es el que m√°s paga en comisiones.
print("El coste total de comisiones del cliente con mas contratos es: {:.2f}".format(coste_cliente_mas_contratos.sum()))
print("El promedio del coste total de comisiones de los dem√°s clientes es: {:.2f}".format(df_coste_cliente.loc[~df_coste_cliente["idcliente"].isin(idcliente_mas_contratos),"imp_comisiones"].mean()))


#                                EJERCICIO 3
#Por cuestiones comerciales se han seleccionado cinco clientes: 12874803, 13067217, 13163055, 13335511, 12900036. Genera un dataframe llamado clientes_vip que contenta los siguientes campos e informaci√≥n para cada cliente:
#‚Ä¢	Edad: [41, 18, 33, 56, 42]
#‚Ä¢	Nombre: [Pepa A., Antonio G., Fernando T., Rebeca S., Laura F.]
#‚Ä¢	Ingresos anuales: [43000, 25000, 55000, 32000, 40000]
import pandas as pd

clientes_vip = pd.DataFrame({"idcliente": [12874803, 13067217, 13163055, 13335511, 12900036], 
                            "Edad": [41, 18, 33, 56, 42], 
                            "Nombre": ["Pepa A.", "Antonio G.", "Fernando T.", "Rebeca S.", "Laura F."],
                            "Ingresos anuales": [43000, 25000, 55000, 32000, 40000]})
#Genera un nuevo dataframe llamado cli_vip_fin que sea el resultado de a√±adir la informaci√≥n de clientes_vip a df_fil y que contenga √∫nicamente los cinco clientes seleccionados.
cli_vip_fin = pd.merge(df_fil, clientes_vip, on='idcliente')
cli_vip_fin["bonificacion"] = cli_vip_fin["Edad"].apply(lambda x: x//10)
print("\n\t******** Mostrando Dataframe del ejercicio 3 **********\n")
print(cli_vip_fin)

```

```python
#Realiza una copia del objeto df y ll√°malo df2_4, sobre este conjunto de datos realiza las siguientes operaciones:
#Crea una variable llamada limite_gr de forma que contenga una versi√≥n categorizada de la variable limite conforme a los siguientes tramos:

#[0, 700]
#(700, 800]
#(800, 1200]
#(1200, 1700]
#(1700, 2500]
#(2500, 5000]
#(5000, 8000]
#(8000, 10000]
#(10000, 20000]
import pandas as pd
import numpy as np
df = pd.read_csv('tarjetas.txt', delimiter='\t', decimal=',')
df2_4 = df.copy()
bins = [0, 700, 800, 1200, 1700, 2500, 5000, 8000, 10000, 20000]
labels = ["[0, 700]", "(700, 800]", "(800, 1200]", "(1200, 1700]", "(1700, 2500]", "(2500, 5000]", "(5000, 8000]", "(8000, 10000]", "(10000, 20000]"]
df2_4["limite_gr"] = pd.cut(df2_4["limite"], bins, labels=labels)

#¬øCu√°ntos registros hay en cada categor√≠a de la variable limite_gr? ¬øCu√°l es el que acumula un mayor porcentaje de registros y qu√© porcentaje es?
value_counts = df2_4['limite_gr'].value_counts()
value_counts_pct = df2_4['limite_gr'].value_counts(normalize=True) * 100

print("Registros en cada categor√≠a\n")
print(value_counts)
print("\nPorcentaje en cada categor√≠a\n")
print(value_counts_pct)

percentages = value_counts / len(df2_4) * 100
max_category = percentages.idxmax()
max_percentage = percentages.max()

print("\nLa categor√≠a con el porcentaje m√°s alto es " + max_category + " con " + str(max_percentage) + "%")

#Selecciona una muestra aleatoria del 10% de los registros de df2_4 y almacen√°la en df2_4_smp. 
#¬øQu√© porcentaje de registros hay en cada categor√≠a de la variable limite_gr en la muestra?
df2_4_smp = df2_4.sample(frac=0.1, random_state=1)
value_counts_smp = df2_4_smp['limite_gr'].value_counts(normalize=True) * 100
print("---------------------------------------------------------------------------------")
print("Muestra aleatoria del 10% con sus porcentajes de registro en cada categoria de la variable limite_gr\n",value_counts_smp)

# Compara ambos porcentajes en el mismo dataframe.
#Calculo los porcentajes en la muestra
df2_4_smp_pct = df2_4_smp['limite_gr'].value_counts(normalize=True) * 100

#Combino ambas tablas de porcentajes
df_pct = pd.concat([value_counts_pct, df2_4_smp_pct], axis=1)
print("\n\t\t**COMBINANDO AMBAS TABLAS DE PORCENTAJES**")
df_pct.columns = ['Porcentaje en todos los registros', 'Porcentaje en la muestra']

#Imprimo el dataframe resultante
print(df_pct)

#Dentro de la empresa se utiliza la cifra
#ùëëùëñùë†ùëùùëúùëõùëñùëèùëôùëí2‚àíùëò‚ãÖùëëùëñùë†ùëùùë¢ùëíùë†ùë°ùëúùëôùëñùëöùëñùë°ùëí2

#como un factor de propensi√≥n a la sobreutilizaci√≥n de la tarjeta. El componente  ùëò  toma un valor que depende del l√≠mite de la operaci√≥n:
#k = 0.5 si limite < 1500
#k = 1 si limite < 5000
#k = 1.2 si limite < 10.000
#k = 1.5 si limite >= 10.000 Crea una columna llamada propension de forma que contenga el c√°lculo anterior realizado a trav√©s de una funci√≥n.

import matplotlib.pyplot as plt
def calcula_propension(row):
    if row['limite'] < 1500:
        k = 0.5
    elif row['limite'] < 5000:
        k = 1
    elif row['limite'] < 10000:
        k = 1.2
    else:
        k = 1.5
    propension = (((row['disponible']**2 - k)*(row['dispuesto']))/(row['limite']**2))
    return propension

df2_4['propension'] = df2_4.apply(calcula_propension, axis=1)

#Dibuja el histograma de la variable propension.
plt.hist(df2_4["propension"])
plt.xlabel("Propension")
plt.ylabel("Frecuencia")
plt.title("Histograma de Propension")
plt.show()


#¬øCu√°ntos registros tienen una propensi√≥n entre 0.95 y 1.05?
registros_entre_0_95_y_1_05 = df2_4.loc[(df2_4["propension"] > 0.95) & (df2_4["propension"] < 1.05)].shape[0]
print("Registros que tienen una propension entre 0.95 y 1.05: ",registros_entre_0_95_y_1_05)

# ¬øCada factor  ùëò  a cu√°ntos registros ha sido aplicado?
def asigna_k(limite):
    if limite < 1500:
        return 0.5
    elif limite < 5000:
        return 1
    elif limite < 10000:
        return 1.2
    else:
        return 1.5

df2_4["k"] = df2_4["limite"].apply(asigna_k)

counts_k = df2_4["k"].value_counts()
print("\n***cantidad de registros tienen cada valor de k***")
print(counts_k)

# La propension permite conocer si hay un alto o bajo riesgo de que el limite de la tarjeta pueda llegar a ser superado.
#Para ello es necesario identificar que registros son de alto riesgo y cu√°les no. 
#Crea una columna llamada riesgo que contenga los valores alto riesgo y bajo riesgo, 
#se considera que un cliente es de alto riesgo si la propensi√≥n supera al porcentaje de uso.

def evaluar_riesgo(propension, porcentaje_uso):
    if propension > porcentaje_uso:
        return "alto riesgo"
    else:
        return "bajo riesgo"

porcentaje_uso = 0.8
df2_4["riesgo"] = df2_4["propension"].apply(lambda x: evaluar_riesgo(x, porcentaje_uso))

#Para priorizar acciones comerciales se desea marcar al 5% de los clientes con menor porcentaje de uso con la etiqueta 
#prioridad 1, al 1% de los clientes con mayor porcentaje de uso como prioridad 4, al 21% de los clientes con
#mayor porcentaje de uso con prioridad 3 y el resto con prioridad 2.
#El porcentaje de uso a utilizar debe ser agregado para cada cliente ¬øcu√°ntos clientes hay en cada grupo de prioridad?
df2_4['porcentaje_uso'] = (df2_4['disponible'] / df2_4['limite']) * 100

bins = [0, df2_4['porcentaje_uso'].quantile(.05),
        df2_4['porcentaje_uso'].quantile(.96),
        df2_4['porcentaje_uso'].quantile(.99),
        100]
labels = ['prioridad 1', 'prioridad 2', 'prioridad 3', 'prioridad 4']

df2_4['prioridad'] = pd.cut(df2_4['porcentaje_uso'], bins=bins, labels=labels)
print("\n***Cantidad de clientes en cada grupo de prioridad***")
print(df2_4['prioridad'].value_counts())


#La empresa ha decidido aplicar una comisi√≥n de un 1,5% a las tarjetas de alto riesgo sobre el importe disponible y una bonificaci√≥n del 0,43% a las tarjetas de bajo riesgo sobre el importe dispuesto.
#‚Ä¢	¬øCu√°l es el importe total recaudado por la empresa con este nuevo importe?

importe_alto_riesgo = df2_4.loc[df2_4['riesgo'] == 'alto riesgo', 'disponible'].sum() * 0.015
importe_bajo_riesgo = df2_4.loc[df2_4['riesgo'] == 'bajo riesgo', 'dispuesto'].sum() * 0.0043
importe_total = importe_alto_riesgo + importe_bajo_riesgo
print("\nEl importe total recaudado por la empresa es: ", importe_total)

# ¬øQu√© coste tiene para la empresa esta nueva bonificaci√≥n?
importe_bajo_riesgo = df2_4[df2_4['riesgo'] == 'bajo riesgo']['disponible'].sum()
coste_bonificacion = importe_bajo_riesgo * 0.0043
print("El coste de la bonificaci√≥n para la empresa es de: ", coste_bonificacion)

# ¬øHa resultado beneficioso para la empresa aplicar esta nueva medida?
importe_comisiones = (df2_4[df2_4['riesgo'] == 'alto riesgo']['disponible'] * 0.015).sum()
importe_bonificaciones = (df2_4[df2_4['riesgo'] == 'bajo riesgo']['dispuesto'] * 0.0043).sum()
beneficio_neto = importe_bonificaciones - importe_comisiones

if beneficio_neto > 0:
    print("La nueva medida ha resultado beneficiosa para la empresa, ya que ha obtenido un beneficio neto de: ", beneficio_neto)
else:
    print("La nueva medida no ha resultado beneficiosa para la empresa, ya que ha obtenido un beneficio neto de: ", beneficio_neto)
```

```python
#La empresa se est√° planteando transformar ciertas tarjetas en pr√©stamos. Para cuando los clientes lo soliciten ser√° necesario disponer del cuadro de amortizaci√≥n resultante de realizar esta transformaci√≥n. Para ello es necesario lo siguiente:
#‚Ä¢	El importe del pr√©stamo ser√° el saldo dispuesto de la tarjeta +10% de comisi√≥n +5% del l√≠mite si la tarjeta no est√° sobredispuesta en este porcentaje.
#‚Ä¢	Se considerar√° un tipo de inter√©s fijo del 2,75%, aunque debe poder elegirse otro.
#‚Ä¢	El plazo amortizaci√≥n ser√° preestablecido en 5 a√±os, pero podr√≠a modificarse.
#‚Ä¢	Se considera sistema de amortizaci√≥n franc√©s (cuota fija) para amortizaci√≥n mensual.

import pandas as pd
import numpy as np
# Datos del pr√©stamo
saldo_dispuesto = 1000  # importe del saldo dispuesto
comision = 0.1   # comisi√≥n del 10%
limite = 5000
porcentaje_sobre = 0.05
tasa_interes = 0.0275  # tasa de inter√©s anual del 2.75%
plazo = 60  # plazo en meses

# C√°lculo del importe a prestar
importe_prestamo = saldo_dispuesto + saldo_dispuesto * comision
if limite * porcentaje_sobre > importe_prestamo:
    importe_prestamo += limite * porcentaje_sobre

#c√°lculo del factor de amortizaci√≥n
factor_amortizacion = (tasa_interes * (1 + tasa_interes) ** plazo) / ((1 + tasa_interes) ** plazo - 1)

#Inicializaci√≥n de listas para almacenar los resultados
cuotas = []
intereses = []
amortizaciones = []
saldos = []

#c√°lculo de las cuotas mensuales
saldo_restante = importe_prestamo
for i in range(plazo):
    interes = saldo_restante * tasa_interes
    amortizacion = factor_amortizacion - interes
    cuota = interes + amortizacion
    saldo_restante -= amortizacion
    cuotas.append(cuota)
    intereses.append(interes)
    amortizaciones.append(amortizacion)
    saldos.append(saldo_restante)
    
#No hice la funci√≥n del cuadro de amortizacion porque no me sali√≥, asi que resolv√≠ el problema de la siguiente manera:
#Creaci√≥n del cuadro de amortizaci√≥n
data = {
'cuota': cuotas,
'interes': intereses,
'amortizacion': amortizaciones,
'saldo': saldos
}

df_amortizacion = pd.DataFrame(data)
 

#Imprimir cuadro de amortizaci√≥n
print("***Cuadro de amortizaci√≥n***\n")
print(df_amortizacion)


#Con las condiciones prefijadas de antemano ¬øcu√°l ser√≠a la cuota inicial que se pagar√≠a por cada tarjeta?
cuota_inicial = factor_amortizacion * importe_prestamo
print("\nLa cuota inicial a pagar es: ",cuota_inicial)

#Esta operaci√≥n solo ser√≠a aprobada si la cuota resultante supone un importe inferior al l√≠mite de la tarjeta dividido entre 55 ¬øqu√© porcentaje de tarjetas es susceptible de ser aprobado?
import random

# l√≠mite de la tarjeta
limite = 5000

# n√∫mero de operaciones a simular
num_operaciones = 1000

# Contador de operaciones aprobadas
operaciones_aprobadas = 0

for i in range(num_operaciones):
    # Generar datos aleatorios para el saldo dispuesto y la comisi√≥n
    saldo_dispuesto = random.randint(500,3000)
    comision = random.uniform(0.05,0.15)
    # C√°lculo del importe a prestar
    importe_prestamo = saldo_dispuesto + saldo_dispuesto * comision
    #c√°lculo del factor de amortizaci√≥n
    factor_amortizacion = (tasa_interes * (1 + tasa_interes) ** plazo) / ((1 + tasa_interes) ** plazo - 1)
    #c√°lculo de la cuota inicial
    interes = importe_prestamo * tasa_interes
    amortizacion = factor_amortizacion - interes
    cuota_inicial = interes + amortizacion
    # Comparar la cuota inicial con el l√≠mite de la tarjeta dividido entre 55
    if cuota_inicial <= limite / 55:
        operaciones_aprobadas += 1

# Calculo el porcentaje de operaciones aprobadas
porcentaje_aprobadas = (operaciones_aprobadas / num_operaciones) * 100
print("El porcentaje de operaciones aprobadas es: " + str(round(porcentaje_aprobadas, 2)) + "%")

#El porcentaje de operaciones aprobadas depender√° del rango de los valores aleatorios generados para el saldo dispuesto 
#y la comisi√≥n. Empero,se puede inferir que el porcentaje de operaciones aprobadas ser√° menor al 55% ya que 
#la cuota inicial siempre ser√° mayor

```

```python
#Programa una funci√≥n que permita aplicar la transformaci√≥n de Box - Cox sobre una columna de un dataframe de Pandas. 
#Selecciona un valor para  ùúÜ  y util√≠zalo para aplicar la transformaci√≥n sobre la variable dispuesto. 
#Apl√≠cale tambi√©n el valor 0.982.
from scipy.stats import boxcox,jarque_bera
import pandas as pd
import numpy as np
from scipy.optimize import minimize_scalar

def boxcox_transform(dataframe, column, lambda_value):
    dataframe[column] = boxcox(dataframe[column], lambda_value)
    return dataframe

# Uso
df = pd.DataFrame({'x': [1, 2, 3, 4, 5]})
df_transformed = boxcox_transform(df, 'x', 0.982)
print(df_transformed)

#Programa una funci√≥n que permita obtener el valor  ùúÜ  √≥ptimo, con una precisi√≥nd de dos decimales, 
#al aplicar la familia de transformaciones Box - Cox sobre la variable dispuesto de forma que la variable 
#se asemeje lo m√°ximo posible a una distribuci√≥n Normal.

def jb_box_cox(x, lmbda):
    y = (x**lmbda - 1)/lmbda
    return jarque_bera(y)[0]

def encontrar_lambda_optim(x):
    res = minimize_scalar(jb_box_cox, args=(x,), bounds=(-1, 1), method='bounded')
    return round(res.x, 2)


x = np.array((1, 23, 4))
lambda_optim = encontrar_lambda_optim(x)
print("\nValor √≥ptimo de ùúÜ: ",lambda_optim)

#Dibuja conjuntamente el histograma de la varible transformada y la variable sin transformar.
import matplotlib.pyplot as plt
y = (x**lambda_optim - 1)/lambda_optim
plt.hist(x, bins=20, color='blue', alpha=0.5, label='x sin transformar')
plt.hist(y, bins=20, color='red', alpha=0.5, label='x transformada') #label es el t√≠tulo que llevar√° en el gr√°fico.
plt.title("Histograma de la variable x")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.legend() #a√±adir leyenda al gr√°fico 
plt.show() #mostrar el gr√°fico




```
